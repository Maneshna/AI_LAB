```python
adj = {
    1: [2], 2: [1, 4], 3: [2], 4: [2]}
n = max(adj.keys())
matrix = [[0] * n for _ in range(n)]
for u in range(1, n+1):
    for v in adj[u]:
        matrix[u-1][v-1] = 1
print("matrix")

for row in matrix:
    print(row)
```

```python
stack = []
while True:
    print("\n--- STACK OPERATIONS ---")
    print("1. Push")
    print("2. Pop")
    print("3. Top element")
    print("4. Display Stack")
    print("5. Exit")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

    if choice == 1:
      value = input("Enter the value to push: ")
      stack.append(value)
      print(f"{value} pushed to stack.")
    elif choice == 2:
      if len(stack) == 0:
        print("Stack is empty.")
      else:
        value = stack.pop()
        print(f"{value} popped from stack.")
    elif choice == 3:
      if len(stack) == 0:
        print("Stack is empty.")
      else:
        print(f"Top element: {stack[-1]}")
    elif choice == 4:
      if len(stack) == 0:
        print("Stack is empty.")
      else:
        print("Stack elements:", stack)
    elif choice == 5:
      print("Exiting program.")
      break
    else:
      print("Invalid choice")
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return "Stack is empty"

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)


# my_stack = Stack()
# print("Is stack empty?", my_stack.is_empty())

# my_stack.push(10)
# my_stack.push(20)
# my_stack.push(30)
# print("Stack after pushes:", my_stack) # [10, 20, 30]
# print("Stack size:", my_stack.size()) # 3

# print("Top element (peek):", my_stack.peek()) # 30

# popped_item = my_stack.pop()
# print("Popped item:", popped_item)
# print("Stack after pop:", my_stack)

# popped_item = my_stack.pop()
# print("Popped item:", popped_item) # 20
# popped_item = my_stack.pop()
# print("Popped item:", popped_item) # 10

# print("Is stack empty?", my_stack.is_empty()) # True
# print("Pop from empty stack:", my_stack.pop()) # Stack is empty
# print("Peek on empty stack:", my_stack.peek()) # Stack is empty
```

```python
#check parethesis
def is_balanced(s):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for char in s:

        if char in '({[':
            stack.append(char)


        elif char in ')}]':
            if not stack:
                return False
            if stack[-1] != pairs[char]:
                return False
            stack.pop()


    return len(stack) == 0



expr = input("Enter parentheses expression: ")

if is_balanced(expr):
    print("Balanced")
else:
    print("Not Balanced")
```

```python
#reverse a string using stack
def reverse_string(s):
    stack = []
    for char in s:
        stack.append(char)
    reversed_str = ""
    while stack:
        reversed_str += stack.pop()
    return reversed_str
string = input("Enter a string: ")
print("Reversed string:", reverse_string(string))
```
