Implement BFS and DFS to solve a maze.
Problem Statement: Given a grid-based maze where 0 represents walls and 1
represents walkable paths, find the shortest path from a start cell to an end cell.
Tasks:
● Use BFS to find the shortest path.
● Use DFS to explore all possible paths and report one valid path (not necessarily
the shortest).
● Compare the number of nodes explored by BFS and DFS.


```python
from collections import deque
```

```python
def reconstruct_path(parent, start, end):
    path = []
    current = end
    while current != start:
        path.append(current)
        current = parent[current]
    path.append(start)
    return path[::-1]
```

```python
def bfs_shortest_path(maze, start, end):
    rows, cols = len(maze), len(maze[0])
    queue = deque([start])
    visited = set([start])
    parent = {}
    nodes_explored = 0

    directions = [(1,0), (-1,0), (0,1), (0,-1)]

    while queue:
        current = queue.popleft()
        nodes_explored += 1

        if current == end:
            return reconstruct_path(parent, start, end), nodes_explored

        r, c = current
        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if (0 <= nr < rows) and (0 <= nc < cols) and maze[nr][nc] == 1:
                if (nr, nc) not in visited:
                    visited.add((nr, nc))
                    parent[(nr, nc)] = current
                    queue.append((nr, nc))

    return None, nodes_explored
```

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

print("\nEnter the maze row by row (0 for wall, 1 for path):")
maze = []
for i in range(rows):
    row = list(map(int, input().split()))
    maze.append(row)

# Start and end coordinates
print("\nEnter start cell (row col):")
sr, sc = map(int, input().split())

print("Enter end cell (row col):")
er, ec = map(int, input().split())

start = (sr, sc)
end = (er, ec)

bfs_path, bfs_nodes = bfs_shortest_path(maze, start, end)

print("BFS Shortest Path:", bfs_path)
print("BFS Nodes Explored:", bfs_nodes)
```
