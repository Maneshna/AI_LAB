Objective: Use Bi-directional BFS/DFS to solve a navigation problem.
Problem Statement: Represent a city map as a graph where intersections are nodes
and roads are edges. Find the shortest path between two locations.
Tasks:
● Implement Bi-directional BFS to minimize the number of nodes explored.
● Compare the performance of Bi-directional BFS with standard BFS and DFS.
● Visualize the search process (e.g., using a library like networkx in Python).


```python
class Graph:
    def __init__(self):
        """Initializes a graph with an empty adjacency list."""
        self.adj_list = {}
        print("Graph initialized with an empty adjacency list.")

    def add_node(self, node_id):

        """Adds a node (intersection) to the graph if it doesn't already exist."""
        if node_id not in self.adj_list:
            self.adj_list[node_id] = []
            print(f"Node {node_id} added.")
        else:
            print(f"Node {node_id} already exists.")

    def add_edge(self, node1_id, node2_id, weight=1):
        """Adds an undirected edge (road) between two nodes with an optional weight.
        If nodes do not exist, they are added first.
        """

        if node1_id not in self.adj_list:
            self.add_node(node1_id)
        if node2_id not in self.adj_list:
            self.add_node(node2_id)

        # Add edge in both directions
        self.adj_list[node1_id].append({"node": node2_id, "weight": weight})
        self.adj_list[node2_id].append({"node": node1_id, "weight": weight})
        print(f"Edge added between {node1_id} and {node2_id} with weight {weight}.")

    def get_neighbors(self, node_id):

        return self.adj_list.get(node_id, [])

    def __str__(self):

        graph_str = "Graph Adjacency List:\n"
        for node, neighbors in self.adj_list.items():
            neighbor_list = ", ".join([f"{n['node']}(w={n['weight']})" for n in neighbors])
            graph_str += f"  {node}: [{neighbor_list}]\n"
        return graph_str

print("Graph class defined with add_node, add_edge, get_neighbors, and __str__ methods.")
```
